```python
%matplotlib inline
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
df = pd.read_csv('train.csv')
df.head()
```

```python
df.isnull().sum()
```

```python
df.replace(['S', 'C', 'Q', np.nan], [1, 2, 3, 0], inplace=True)
```

```python
df.replace(['male', 'female'], [1, 2], inplace=True)
```

```python
df = df.drop(df.columns[[0, 3, 8, 10]], axis=1)
sns.pairplot(df, hue="Survived")
```
